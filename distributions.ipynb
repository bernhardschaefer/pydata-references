## Exponential distribution

```python
def p(x, λ):
    return λ*np.exp(-λ*x)

x = np.linspace(0,5,1000)

fig, ax = plt.subplots(figsize=(10,3))
for λ in [.2, 1., 5.]:
    ax.plot(x, p(x,λ), label=λ)
ax.legend()
ax.set_ylim(0, 2)
ax.set_xlim(0, 5);
```

## Power-law distribution

```python
def p(x, α, c=1):
    return c*x**-α

x = np.linspace(1e-5, 5, 1000)

fig, ax = plt.subplots(figsize=(5,5))
for α in [1, 3, 5]:
    ax.plot(x, p(x,α), label=α)
ax.legend()
ax.set_ylim(0, 5)
ax.set_xlim(0, 5);
```
