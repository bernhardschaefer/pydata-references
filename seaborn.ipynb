# Reference for plotting with seaborn

- This is a work in progress
- The train.csv is based on the Ames Housing dataset and taken from Kaggle: https://www.kaggle.com/c/house-prices-advanced-regression-techniques

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline
pd.set_option('display.max_rows', 100)
pd.set_option('display.max_columns', 100)
```

```python
df = pd.read_csv('data/train.csv')
df.head()
```

```python
df.nunique().to_frame().transpose()
```

## Univariate Distributions

### Continuous Variable

```python
from scipy.stats import lognorm
ax = sns.distplot(df.SalePrice, fit=lognorm, kde=True)
```

```python
fig, (ax0, ax1, ax2) = plt.subplots(ncols=3, nrows=1, figsize=(16,6))

sns.distplot(df.LotArea, ax=ax0)

sns.distplot(np.log(df.LotArea), ax=ax1)

# needs a lot of bins to look similar to prior log-transform since bins are getting smaller to the right
ax2.set_xscale('log')
sns.distplot(df.LotArea, bins=500, ax=ax2)
```

### Categorical Variable

```python
ax = sns.countplot(df.MSSubClass)
```

## Bivariate

### Two continuous variables

- Scatter plot only applicable to small datasets, otherwise leads to overplotting.
- Seaborn joint probability distribution plot includes:
  - scatter plot
  - marginal histograms
  - correlation + p-value
  - [optional] kind of plot to draw, e.g. regression

```python
sns.jointplot(np.log(df.LotArea), df.SalePrice, kind="reg")
```

```python
sns.lmplot(x="LotArea", y="SalePrice", data=df[df.LotArea < 30000])
```

### One categorical, one continuous variable

- swarmplot/stripplot works well for small datasets
- violinplot/boxplot for larger datasets

```python
# plt.figure for changing figsize of single plot
plt.figure(figsize=(18,6))
ax = sns.swarmplot(x="MSSubClass", y="SalePrice", data=df)
```

```python
plt.figure(figsize=(18,6))
ax = sns.violinplot(x="MSSubClass", y="SalePrice", data=df)
```

```python
plt.figure(figsize=(18,6))
ax = sns.boxplot(x="MSSubClass", y="SalePrice", data=df)
```

### Two categorical variables

```python
plt.figure(figsize=(18,6))
ax = sns.countplot(y="MSZoning", hue="MSSubClass", data=df)
```
